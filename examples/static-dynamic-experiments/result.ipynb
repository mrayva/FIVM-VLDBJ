```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter, StrMethodFormatter

# Sample data
times = np.array([38152, 3843])
N = 84055817
values = N / times * 1000  # tuples per second
categories = ['All Dynamic', 'Mixed']  # Labels for the bars

print(values)

# Create figure and axis
plt.rcParams.update({'font.size': 22})  # Set global font size to 22
fig, ax = plt.subplots(figsize=(6, 6))  # Slightly larger figure to accommodate larger text

# Create the bar chart with narrower bars
bars = ax.bar(categories, values, color='#3498db')

# Add value labels on top of each bar with scientific notation
for bar in bars:
    height = bar.get_height()
    formatted_value = f'{height/1e6:.2f}M'  # Format as millions with 2 decimal places
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.02*max(values),
            formatted_value, ha='center', va='bottom', fontsize=20)

# Customize the chart
ax.set_ylabel('Throughput (tuples per second)', fontsize=22)
ax.set_ylim(0, max(values) * 1.15)  # Set y-axis limit with some padding

# Format y-axis to use millions with M suffix
def millions(x, pos):
    return f'{x/1e6:.0f}M'

ax.yaxis.set_major_formatter(plt.FuncFormatter(millions))

# Increase tick label size
ax.tick_params(axis='both', which='major', labelsize=20)

# Add grid lines for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add a light background color to make the chart more visually appealing
ax.set_facecolor('#f8f8f8')

# Remove top and right spines for cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the plot
plt.tight_layout()
plt.show()
```
